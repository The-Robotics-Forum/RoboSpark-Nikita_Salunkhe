```python
df2.head()
```

```python
sns.jointplot("Marketing Spend", "Profit", data = df2, xlim =(100000, 500000), ylim =(10000, 200000), height = 10, alpha = 0.7, marker = '+')
# Using seaborn library to plot in an better UI 
# height = graph image height
# marker = plotting style
# alpha = transperancy of our graph
```
