```python
def f(x) : # square function for plotting parabola
    f = x ** 2
    return f
def Df(x) : # derivative function of f(x)
    return 2*x
def update(x, alpha): # updating value of x and alpha is it's learning rate
    up = x - alpha * Df(x)
    return up
```

```python
x = np.linspace(-5, 5, 100) # placing x linearly
x
```

```python
def plot_s(guess, alpha, epoch) : #epoch = number of iteration
    fig, ax = plt.subplots() #plotting graph
    x = np.linspace(-5, 5, 100) # placing x linearly
    ax.plot(x, f(x)) # plotting the parabola
    x = guess
    ax.plot(x, f(x), 'o', label='start x=%.2f' %x) 
    for i in range(epoch):
        xold = x
        x = update(x, alpha) # updating new value to function
        ax.plot(x, f(x), 'o', label='start x=%.2f' %x) # plotting the points
        ax.plot([xold, x], [f(xold), f(x)], '-') # for connecting line between two points
    plt.legend()
```

```python
plot_s(-2.5, 1.05, 10)
```

```python
plot_s(-2.5, 0.9, 10)
```

```python
plot_s(-2.5, 0.02, 10)
```

```python
plot_s(-2.5, 0.5, 10)
```
