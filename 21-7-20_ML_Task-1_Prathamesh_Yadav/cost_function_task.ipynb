```python
def f(x) :                         
    f = x ** 2
    return f
def Df(x):  
    return 2*x
def update(x, alpha):       
    up = x - alpha * Df(x)
    return up
```

```python
x = np.linspace(-5, 5, 100)
def plot_s(guess, alpha, epoch):
    fig, ax = plt.subplots()
    x = np.linspace(-5, 5, 100)
    ax.plot(x, f(x))
    x = guess
    ax.plot(x, f(x), 'o', label='start x=%.2f' %x)
    for i in range(epoch):
        xold = x
        x = update(x, alpha)
        ax.plot(x, f(x), 'o', label='x = %.2f' %x)
        ax.plot([xold, x], [f(xold), f(x)], '-')
    plt.legend()
```

```python
plot_s(-2, 0.5, 10)
```

```python
plot_s(-4, 0.05, 10)
```
